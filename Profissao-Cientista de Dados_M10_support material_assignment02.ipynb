1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats

# Carregando os dados
data = pd.read_csv('previsao_de_renda.csv')

# Boxplot para 'posse_de_imovel'
plt.figure(figsize=(10, 6))
sns.boxplot(x='posse_de_imovel', y='renda', data=data)
plt.title('Distribuição de Renda por Posse de Imóvel')
plt.xlabel('Posse de Imóvel')
plt.ylabel('Renda')
plt.show()

# Boxplot para 'posse_de_veiculo'
plt.figure(figsize=(10, 6))
sns.boxplot(x='posse_de_veiculo', y='renda', data=data)
plt.title('Distribuição de Renda por Posse de Veículo')
plt.xlabel('Posse de Veículo')
plt.ylabel('Renda')
plt.show()

# Teste estatístico para 'posse_de_imovel'
renda_imovel = data[data['posse_de_imovel'] == True]['renda']
renda_sem_imovel = data[data['posse_de_imovel'] == False]['renda']
t_stat, p_val = stats.mannwhitneyu(renda_imovel, renda_sem_imovel, alternative='two-sided')
print(f"Teste de Mann-Whitney para Posse de Imóvel: p-valor = {p_val:.4f}")

# Teste estatístico para 'posse_de_veiculo'
renda_veiculo = data[data['posse_de_veiculo'] == True]['renda']
renda_sem_veiculo = data[data['posse_de_veiculo'] == False]['renda']
t_stat, p_val = stats.mannwhitneyu(renda_veiculo, renda_sem_veiculo, alternative='two-sided')
print(f"Teste de Mann-Whitney para Posse de Veículo: p-valor = {p_val:.4f}")
````

- O Teste de Mann-Whitney complementa essa análise visual ao testar se as diferenças nas medianas de renda são estatisticamente significativas. Se o p-valor deste teste for menor que 0,05, concluímos que as diferenças são significativas, reforçando a evidência de que a posse de imóveis ou veículos pode ser um bom preditor de renda mais alta. Esses resultados juntos ajudam a identificar fatores que influenciam a renda, possibilitando uma melhor segmentação de clientes ou ajuste de produtos financeiros como limites de crédito.
